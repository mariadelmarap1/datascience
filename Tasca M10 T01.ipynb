```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
```

```python
#Exercici 1: 
#Realitza web scraping de dues  pàgines web  utilitzant BeautifulSoup primer i Selenium després. 
#A1: pagina 1- Utilitzant BeautifulSoup
URL = "https://quotes.toscrape.com/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find(class_="quote")
print(results.prettify())
```

```python
# Encontrar todos los elementos de quotes
quote_elements = soup.find_all('div', {'class': 'quote'})
print(quote_elements)
```

```python
# Iterar sobre cada elemento de quote
for quote_element in quote_elements:
    # Encontrar el elemento de texto
    text_element = quote_element.find('span', {'class': 'text'})
    text = text_element.text.strip()
    print(text)

    # Encontrar el elemento de autor
    author_element = quote_element.find('small', {'itemprop': 'author'})
    author = author_element.text.strip()
    print(author)

    # Encontrar los elementos de etiqueta
    tag_elements = quote_element.find_all('a', {'class': 'tag'})
    tags = [tag.text for tag in tag_elements]
    print(tags)
```

```python
#Extract links from tag_elements
tag_set = set()

for quote_element in quote_elements:
    tag_elements = quote_element.find_all('a', {'class': 'tag'})
    for tag in tag_elements:
        tag_text = tag.text
        tag_link = tag.get('href')
        if tag_text not in tag_set:
            print(tag_text)
            tag_set.add(tag_text)
        print(tag_link)
```

```python
#Exercici 2 i 3: Tria una pàgina web que tu vulguis i realitza web scraping mitjançant la llibreria Selenium, graba en un doc de word los resultados
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from docx import Document


# Configurar el driver de Selenium
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  # Abre la ventana maximizada
driver = webdriver.Chrome(options=options)

# Inicializar el driver de Selenium
driver = webdriver.Chrome()

# Navegar a la página
driver.get('https://www.fotocasa.es/es/comprar/casas-rurales/barcelona-provincia/todas-las-zonas/l?maxPrice=225000&minPrice=200000&minRooms=4')
time.sleep(10)

# Encontrar los elementos que contienen los nombres de casa y los precios
nombres_de_casa = driver.find_elements(By.XPATH, '//span[@class="re-CardTitle"]')
precios = driver.find_elements(By.XPATH, '//span[@class="re-CardPrice"]')

# Iterar a través de los elementos y obtener el texto
for i in range(len(nombres_de_casa)):
    nombre = nombres_de_casa[i].text
    precio = precios[i].text
    print(nombre, precio)
    
# Crear el documento de Word
doc = Document()
doc.add_heading('Resultados de scraping', 0)

# Crear la tabla y agregar las filas con los datos de scraping
tabla = doc.add_table(rows=1, cols=4)
tabla.style = 'Table Grid'
encabezados = tabla.rows[0].cells
encabezados[0].text = 'Rk'
encabezados[1].text = 'Categoría'
encabezados[2].text = 'Lugar'
encabezados[3].text = 'Valor'
for i in range(len(nombres_de_casa)):
    nombre = nombres_de_casa[i].text
    precio = precios[i].text
    fila = tabla.add_row().cells
    fila[0].text = str(i+1)
    fila[1].text = 'Finca Rústica'
    fila[2].text = nombre
    fila[3].text = precio

# Guardar el documento
doc.save('resultados_scraping.docx')

# Cerrar el navegador
driver.quit()
```
