```python
import pandas as pd
import numpy as np
```

```python
#Exercici 1: Descarrega el data set Airlines Delay (muestra de 20), Explora les dades (columnas) que conté

archivo = pd.read_csv('DelayedFlights.csv')
archivo[:20]
```

```python
#Explora les dades (columnas) que conté.
#El archivo cuenta con los siguientes columnas:

archivo.columns
```

```python
#Exercici 1: queda’t únicament amb les columnes que consideris rellevants. Justifica la teva elecció
#Respuesta: 
#Columnas que no considero relevantes: razón por la cual no me parecen relevantes 
#Year: todos son 2008
#Cancelled: Lo que estamos analizando son retrasos, no si se canceló o no. 
#Unamed: No me aporta información y clsifica los datos por números, que ya estaban clasificados. 
#Codigo de cancelación: No estoy analizando cancelación sino retraso 

newarchivo = archivo.drop(['Year', 'Unnamed: 0', 'Cancelled', 'CancellationCode'], axis='columns')
```

```python
#Exercici 1: Redueix la dimensió del dataset de manera aleatòria per tal d'obtenir un dataset de només 200.000 registres. Tots els exercicis s'han de fer amb aquest dataset reduït.

aarchivo = newarchivo.sample(n=200000)
aarchivo[:20]
```

```python
#Exercici 2: Resumeix estadísticament el dataset i les columnes d’interès. 
#Fes una anàlisi estadístic del que consideris rellevant.

archivo2 = aarchivo.drop(['Month', 'DayofMonth', 'DayOfWeek', 'FlightNum', 'TailNum' , 'Origin' , 'Dest' , 'Diverted'], axis='columns')
archivo2.describe()
```

```python
#Exercici 2: Troba quantes dades faltants hi ha per columna

aarchivo.isna().sum() 
```

```python
#Exercici 2: Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

barchivo = aarchivo.assign(VelAir=np.nan, YesNoDelay='yes')
print (barchivo)
```

```python
#Exercici 2: Fes una taula de les aerolínies amb més endarreriments acumulats:
# primer paso: filtro el archivo para tener solo los datos que me interesan 

completo = aarchivo[["UniqueCarrier","ArrDelay"]]
print(completo)
```

```python
#segundo paso: Filtro el archivo para tener solo las aerolineas con retrasos

soloretrasos = completo[completo['ArrDelay']>0]
print (soloretrasos)
```

```python
#Tercer paso: Agrupo por aerolineas los retrasos

Agruparetrasos = soloretrasos.groupby(['UniqueCarrier']).count()
print(Agruparetrasos)
```

```python
#Cuarto paso: Pongo de mayor a menor las aerolineas con más retrasos acumulados

print(Agruparetrasos.sort_values(by=['ArrDelay'], ascending=False))
```

```python
#Quins són els vols més llargs? 
#Primer paso: Extraigo las variables que me interesan

vueloslargos = aarchivo[["FlightNum","AirTime"]]
print(vueloslargos)
```
