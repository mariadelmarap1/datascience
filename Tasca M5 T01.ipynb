```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("Lligues europees UEFA.csv", sep=';', header=None, names=['Rk', 'Squad', 'Country', 'LgRk', 'MP', 'W', 'D', 'L' , 'GF' ,'GA' ,'GD' ,'Pts' ,'Pts/G' ,'xG' ,'xGA' ,'xGD' ,'xGD/90' ,'Last 5' ,'Attendance' ,'Top Team Scorer' ,'Scorer' ,'Goalkeeper'], encoding='latin-1')
df[:20]
```

```python
#borro la primera fila, por que eran palabras, y me impedia hacer los calculos estadísticos.
df = df.drop(0)
```

```python
#para hacer los calculos estadísticos, necesito saber si hay valores Nan, y poder borrarlos, en caso de que hayan.
nan_values = df['GF'].isnull().sum()
print("Number of NaN values in column GF:", nan_values)
```

```python
#para hacer los calculos estadísticos, necesito cambiar los valores str a float: es decir numerico
df['GF'] = df['GF'].astype(float)
```

```python
#Exercici 1
#Descarrega el dataset adjunt de dades oficials de la UEFA i selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%. 
#Mi hipótesis nula de que la media de los goles por partido (GF) es igual a 4, hipotesis alternativa, es que el valor de los goles es diferente a 4
#primero es necesario establecer la hipótesis nula y la hipótesis alternativa. 
#Luego, elegir el atributo relevante (GF: GOLES) y realizar una prueba de hipótesis apropiada, dependiendo de la distribución y el tipo de datos.
#A continuación, calculo el p-valor y decido si se rechaza o no la hipótesis nula, utilizando una prueba t-student 




alpha = 0.05

n = len(df['GF'])
mean_GF = df['GF'].mean()
std_GF = df['GF'].std()
t_statistic, p_value = stats.ttest_1samp(df['GF'], 4)

if p_value < alpha:
    print("Se rechaza la hipótesis nula, con un p-valor de", p_value)
else:
    print("No se rechaza la hipótesis nula, con un p-valor de", p_value)
```

```python
#Exercici 2
#Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. 
#Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%
#Hipotesis nula: España y alemania tienen la misma cantidad de goles, estadísticamente. 
#Hipotesis alternativa: España y alemania tienen diferente cantidad de goles, estadísticamente. 

alpha = 0.05

df_esp = df[df['Country'] == 'ESP']
df_ger = df[df['Country'] == 'GER']

n_esp = len(df_esp['GF'])
mean_esp = df_esp['GF'].mean()
std_esp = df_esp['GF'].std()

n_ger = len(df_ger['GF'])
mean_ger = df_ger['GF'].mean()
std_ger = df_ger['GF'].std()

t_statistic, p_value = stats.ttest_ind(df_esp['GF'], df_ger['GF'], equal_var=False)

if p_value < alpha:
    print("Se rechaza la hipótesis nula, con un p-valor de", p_value) #significa que la hipótesis de que los países ESP y GER tienen la misma cantidad de goles es rechazada, pero no me dcie cual tiene mas goles, para eso tendria que comparar las medias. 
else:
    print("No se rechaza la hipótesis nula, con un p-valor de", p_value) #Significa que con un nivel de confianza del 95%
    #(ya que alpha es 0.05). El p-valor indica que la hipótesis nula es verdadera.
```
